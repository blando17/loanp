**Loan Status Prediction**

```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

**Data collection and preprocessing**

```python
loan_dataset = pd.read_csv('/content/train_u6lujuX_CVtuZ9i (1).csv')
```

```python
loan_dataset.head()
```

```python
loan_dataset.isnull().sum()
```

```python
loan_dataset = loan_dataset.dropna()
```

```python
loan_dataset.isnull().sum()
```

Encoding yes no to binary

```python
loan_dataset.replace({"Loan_Status":{'N':0,'Y':1}},inplace=True)
```

```python
loan_dataset.head()
```

```python
loan_dataset['Dependents'].value_counts()
```

Replacing 3+ to 4

```python
loan_dataset = loan_dataset.replace(to_replace='3+', value=4)
```

```python
loan_dataset['Dependents'].value_counts()
```

**convert categorical columns to numerical values**

```python
loan_dataset.replace({'Married':{'No':0,'Yes':1},'Gender':{'Male':1,'Female':0},'Self_Employed':{'No':0,'Yes':1},
                      'Property_Area':{'Rural':0,'Semiurban':1,'Urban':2},'Education':{'Graduate':1,'Not Graduate':0}},inplace=True)
```

```python
loan_dataset.head()
```

**separating data and target**

```python
x = loan_dataset.drop(columns=['Loan_ID','Loan_Status'],axis=1)
y = loan_dataset['Loan_Status']
print(x)
print(y)
```

Splitting into Training and testing data

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1, stratify=y, random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

Training the Model(SVM)

```python
classifier = svm.SVC(kernel='linear')
```

```python
classifier.fit(x_train,y_train)
```

**Model evaluation**

**on training data**

```python
x_train_prediction = classifier.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction,y_train)
print('Accuracy on training data:', training_data_accuracy)
```

**on testing data**

```python
x_test_prediction = classifier.predict(x_test)
testing_data_accuracy = accuracy_score(x_test_prediction,y_test)
print('Accuracy on testing data:', testing_data_accuracy)
```

**Building a predictive model**

```python

input_data =(0,0,0,1,1,4583,0.0,133.0,360.0,0.0,1)
input_data_as_numpy_array = np.asarray(input_data) #changing the input_data to a numpy array
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)
prediction = classifier.predict(input_data_reshaped)
print(prediction)
if(prediction[0]== 0):
  print("There is No loan")
else:
  print("There is Loan")
```
